# 1. Vectors 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 0. Draw Vectors in Matplotlib

Use `plt.quiver(X, Y, U, V, **kwargs)` to plot vectors in Matplotlib.  
- **X, Y** define the starting positions of the arrows.  
- **U, V** specify the vector components (directions).  

The arguments **X, Y, U, V** are positional-only.

Useful Parameters:
- `angles='xy'` → interprets arrow directions in data coordinates.  
- `scale_units='xy'` → interprets arrow lengths in data coordinates.  
- `scale=1` → disables automatic scaling so arrows match the true vector components.




## 1. What is a Vector

```python
# Example vectors in 2D
v1 = np.array([2, 2])
v2 = np.array([-2, 1])
v3 = np.array([0, 4])

print("Vector 1:", v1)
print("Vector 2:", v2)
print("Vector 3:", v3)

# Visualization
fig, ax = plt.subplots(figsize=(10, 8))

# Draw Vectors from origin
origin = np.array([0, 0])
ax.quiver(origin[0], origin[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005, label=f'v1 = {v1}')#
ax.quiver(origin[0], origin[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005, label=f'v2 = {v2}')
ax.quiver(origin[0], origin[1], v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005, label=f'v3 = {v3}')

ax.set_xlim(-5, 5)
ax.set_ylim(-2, 5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xticks(np.arange(-5, 5, 1))
ax.set_yticks(np.arange(-2, 5, 1))
ax.set_title('2D Vectors Visualization')
ax.grid()
ax.legend()
ax.set_aspect('equal')
ax.axhline(0, color='black',linewidth=0.5, ls='-')
ax.axvline(0, color='black',linewidth=0.5, ls='-')
plt.tight_layout()
plt.show()
```

## 2. Vector Addition 

```python
v1 = np.array([3, 1])
v2 = np.array([1, 2])

v3 = v1 + v2
print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v1 + v2 = {v3}")

# Visualization
fig, ax = plt.subplots(figsize=(10, 8))

# Draw Vectors from origin
origin = np.array([0, 0])
ax.quiver(origin[0], origin[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005, label=f'v1 = {v1}')#
ax.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005, label=f'v2 = {v2}')
ax.quiver(origin[0], origin[1], v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005, label=f'v1 + v2 = {v3}')

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 4)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xticks(np.arange(-1, 5, 1))
ax.set_yticks(np.arange(-1, 5, 1))
ax.set_title('Vector Addition')
ax.grid()
ax.legend()
ax.set_aspect('equal')
ax.axhline(0, color='black',linewidth=0.5, ls='-')
ax.axvline(0, color='black',linewidth=0.5, ls='-')
plt.tight_layout()
plt.show()
```

## 3. Vector Scalar Multiplication 

```python
# Example: 2 * v, 0.5 * v, -1 * v
v = np.array([2, 1])

v_doubled = 2 * v      # Double the length
v_half = 0.5 * v       # Half the length
v_flipped = -1 * v     # Same length, opposite direction

print(f"Original vector: v = {v}")
print(f"2 * v = {v_doubled}  (double length)")
print(f"0.5 * v = {v_half}  (half length)")
print(f"-1 * v = {v_flipped}  (flipped direction)")

# Visualize
fig, ax = plt.subplots(figsize=(10, 8))

vectors = [
    (v, 'b', f'v = {v}'),
    (v_doubled, 'r', f'2v = {v_doubled}'),
    (v_half, 'g', f'0.5v = {v_half}'),
    (v_flipped, 'orange', f'-v = {v_flipped}'),
]

for vec, color, label in vectors:
    ax.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1,
              color=color, width=0.006, label=label)

ax.set_xlim(-3, 5)
ax.set_ylim(-2, 3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Scalar Multiplication')
ax.grid(True)
ax.legend()
ax.set_aspect('equal')
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()

```

## 4. Vector Magnitude

The magnitude (or length) of a vector $v = \begin{bmatrix} v_1 \\ v_2 \end{bmatrix}$ is calculated using the formula: 
$$
||v|| = \sqrt{v_1^2 + v_2^2}
$$

With Numpy you can compute the magnitude of a vector using `np.linalg.norm()` function. 
- **linalg** linear algebra module
- **norm** function to compute vector magnitude

```python
# Example: magnitude of vectors
v1 = np.array([3, 2])
v2 = np.array([5, 0])
v3 = np.array([1, 1])

# Calculate magnitude
mag1 = np.linalg.norm(v1)
mag2 = np.linalg.norm(v2)
mag3 = np.linalg.norm(v3)

# Draw Vectors from origin
fig, ax = plt.subplots(figsize=(8,8))

origin = np.array([0, 0])
vectors = [(v1, 'b', mag1), (v2, 'r', mag2), (v3, 'g', mag3)]
for vector, color, magnitude in vectors:
  ax.quiver(origin[0], origin[1], vector[0], vector[1], color=color, scale=1, scale_units='xy', angles='xy', label=f'{vector}, ||v|| = {magnitude:.2f}')

ax.set_xlim(-1, 6)
ax.set_ylim(-1, 3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xticks(np.arange(-1, 6, 1))
ax.set_yticks(np.arange(-1, 3, 1))
ax.set_title('Vector Magnitudes')
ax.grid()
ax.legend()
ax.set_aspect('equal')
ax.axhline(0, color='black',linewidth=0.5, ls='-')
ax.axvline(0, color='black',linewidth=0.5, ls='-')
plt.tight_layout()
plt.show()
```

## 5. Dot Product
The dot product of two vectors $\underline a = \begin{bmatrix} a_1 \\ a_2 \end{bmatrix}$ and $\underline b = \begin{bmatrix} b_1 \\ b_2 \end{bmatrix}$ is calculated using the formula:
$$
\underline a \cdot \underline b = a_1 b_1 + a_2 b_2 = ||\underline a|| \cdot ||\underline b|| \cdot \cos(\theta)
$$

The dot product is also a projection of one vector onto another.

The **scalar projection** of vector $\underline a$ onto vector $\underline b$ is given by:
$$
proj_{\underline b} (\underline a) = ||a|| \cdot \cos(\theta) = \frac{\underline a \cdot \underline b}{||\underline b||^2}
$$

The **vector projection** of vector $\underline a$ onto vector $\underline b$ is given by:
$$
Proj_{\underline b} (\underline a) = proj_{\underline b} (\underline a) \cdot \underline{b}
$$


```python
a = np.array([2, 2])
b = np.array([5, 1])

# Dot Product
dot = np.dot(a, b) # or dot = a @ b
print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"Dot product: {dot}")

# Projection of a onto b
proj_scalar = np.dot(a, b) / np.linalg.norm(b)**2
proj_vector = proj_scalar * b

print(f"Projection of a onto b: {np.round(proj_vector, 1)}")

dot = proj_scalar * np.linalg.norm(b)**2

print(f"Dot product using projection = scalar projection * ||b||^2 = {dot}")


# Visualization
fig, ax = plt.subplots(figsize=(10, 8))

# Draw Vectors from origin
origin = np.array([0, 0])
ax.quiver(origin[0], origin[1], a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005, label=f'a = {a}')
ax.quiver(origin[0], origin[1], b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005, label=f'b = {b}')

ax.quiver(origin[0], origin[1], proj_vector[0], proj_vector[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005, label=f'Projection of a onto b')#
ax.plot([a[0], proj_vector[0]], [a[1], proj_vector[1]], 'k--', label='Perpendicular Component')

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 4)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xticks(np.arange(-1, 5, 1))
ax.set_yticks(np.arange(-1, 5, 1))
ax.set_title('Vector Addition')
ax.grid()
ax.legend()
ax.set_aspect('equal')
ax.axhline(0, color='black',linewidth=0.5, ls='-')
ax.axvline(0, color='black',linewidth=0.5, ls='-')
plt.tight_layout()
plt.show()
```

## 6. Cross Product
The cross product of two 3D vectors $\underline a = \begin{bmatrix} a_1 \\ a_2 \\ a_3 \end{bmatrix}$ and $\underline b = \begin{bmatrix} b_1 \\ b_2 \\ b_3 \end{bmatrix}$ is calculated using the formula:
$$
\underline a \times \underline b = \begin{bmatrix} a_2 b_3 - a_3 b_2 \\ a_3 b_1 - a_1 b_3 \\ a_1 b_2 - a_2 b_1 \end{bmatrix}
$$

c is a vector that is perpendicular to both $\underline a$ and $\underline b$.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.cross(a, b)

print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"Cross product a x b: {c}")

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Draw Vectors from origin
origin = np.array([0, 0, 0])
ax.quiver(origin[0], origin[1], origin[2], a[0], a[1], a[2], linewidth=2, color='blue', label=f'a = {a}')
ax.quiver(origin[0], origin[1], origin[2], b[0], b[1], b[2], linewidth=2, color='red', label=f'b = {b}')
ax.quiver(origin[0], origin[1], origin[2], c[0], c[1], c[2], linewidth=2, color='green', label=f'a x b = {c}')

ax.set_xlim(0, 8)
ax.set_ylim(0, 8)
ax.set_zlim(0, 8)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Cross Product Visualization: a, b, and a × b")
ax.grid()
ax.legend()
ax.set_box_aspect([1,1,1])
plt.tight_layout()
plt.show()
```
