# 2. Gradients

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Gradient Visualization

```python
# Example f(x, y) = x² + y²
def f(x, y):
  return x**2 + y**2

# Gradient of the scalar function f(x, y) gives a vector field
def gradient_f(x, y):
  dfdx = 2 * x
  dfdy = 2 * y
  return np.array([dfdx, dfdy])

def mag_gradient_f(x, y):
  grad = gradient_f(x, y)
  return np.linalg.norm(grad)

# Check at different points+
points = [(0, 0), (1, 0), (1, 1), (2, 0), (2, 2)]
print("Point (x, y) | Function | Gradient | Gradient Magnitude")
print("-" * 50)
for x, y in points:
  z = f(x, y)
  grad = gradient_f(x, y)
  mag_grad = mag_gradient_f(x, y)
  print(f"({x}, {y})      | f = {z:3d} | [{grad[0]:3.1f}, {grad[1]:3.1f}] | |∇f| = {mag_grad:.2f}")

# 3D Surface
x = np.linspace(-3, 3, 40)
y = np.linspace(-3, 3, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Gradient Vectors
x_grad = np.linspace(-3, 3, 10)
y_grad = np.linspace(-3, 3, 10)
X_grad, Y_grad = np.meshgrid(x_grad, y_grad)
gradients = gradient_f(X_grad, Y_grad)
U = gradients[0]
V = gradients[1]

fig = plt.figure(figsize=(14, 6))

# Visualization 3D Surface
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surface = ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax1.set_title('Surface Plot of f(x, y) = x² + y²')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')

# Visualization Gradient Field
ax2 = fig.add_subplot(1, 2, 2)


contourf = ax2.contourf(X, Y, Z, levels=20, cmap='viridis', alpha=0.6)
ax2.quiver(X_grad, Y_grad, U, V, color='black', scale=7, scale_units='xy', angles='xy', width=0.003)


ax2.set_title('Gradient Vector Field of f(x, y) = x² + y²')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim(-3, 3)
ax2.set_ylim(-3, 3)
ax2.axhline(0, color='black',linewidth=0.5, ls='-')
ax2.axvline(0, color='black',linewidth=0.5, ls='-')
ax2.grid()
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()

```
